# завдання 1 через словник


```python
teams = []

n = int(input("Скільки команд ввести? "))
for i in range(n):
    print(f"\nКоманда №{i + 1}:")
    name = input("Назва команди: ")
    losses = int(input("Кількість поразок: "))
    wins = int(input("Кількість перемог: "))
    draws = int(input("Кількість нічиїх: "))

    team = {
        "назва": name,
        "поразки": losses,
        "перемоги": wins,
        "нічиї": draws
    }
    teams.append(team)

max_team = max(teams, key=lambda t: t["перемоги"])
print("\nКоманда з найбільшою кількістю перемог:", max_team["назва"])
```

# завдання 1 через клас

```python
class Team:
    pass


teams = []

n = int(input("Скільки команд ввести? "))
for i in range(n):
    print(f"\nКоманда №{i + 1}:")
    name = input("Назва команди: ")
    losses = int(input("Кількість поразок: "))
    wins = int(input("Кількість перемог: "))
    draws = int(input("Кількість нічиїх: "))

    team = Team()
    team.назва = name
    team.поразки = losses
    team.перемоги = wins
    team.нічиї = draws

    teams.append(team)

max_team = max(teams, key=lambda t: t.перемоги)
print("\nКоманда з найбільшою кількістю перемог:", max_team.назва)
```
